# Feature Selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
ordered_rank_features= SelectKBest(score_func=chi2,k=10)
ordered_feature=ordered_rank_features.fit(x,y)
```

```python
ordered_rank_features
```

```python
dfscore=pd.DataFrame(ordered_feature.scores_,columns=['score'])
dfcolumns=pd.DataFrame(x.columns)
```

```python
feature_rank=pd.concat([dfcolumns,dfscore],axis=1)
feature_rank.columns=['features', 'score']
feature_rank.sort_values(by='score', ascending=False)
```

```python
feature_rank.nlargest(10,'score')
```

# Feature Importance

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model=ExtraTreesClassifier()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
ranked_feature=pd.Series(model.feature_importances_,index=x.columns)
ranked_feature.nlargest(10).plot(kind='barh')
plt.show()
```

# Information Gain

```python
from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(x, y)
mutual_data=pd.Series(mutual_info, index=x.columns)
mutual_data.sort_values(ascending=False)
```

```python
data=data.drop(['wifi','touch_screen', 'three_g', 'talk_time', 'sc_w', 'dual_sim', 'four_g', 'int_memory', 'blue', 'n_cores', 'mobile_wt', 'm_dep', 'fc'],axis=1)
```

```python
data.head()
```

```python
x=data.drop('price_range', axis=1)
y=data['price_range']
```

```python
print(x.shape)
print(y.shape)
```

# Split the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train= sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

```python
x_train
```

# Using Logistic Regression Algo

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
acc1=accuracy_score(y_test, y_pred)
print(f"Accuracy score:{acc1}")
```

# Using Suport Vector Machine(SVM)

```python
from sklearn.svm import SVC
classifier=SVC(kernel="linear", random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
acc2=accuracy_score(y_test,y_pred)
print(f"Accuracy score:{acc2}")
```

# Using K-NN Algo

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5, metric='minkowski',p=2)
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc3=accuracy_score(y_test, y_pred)
```

```python
print(f"Accuracy score:{acc3}")
```

#  Using Naive Bayes Algo

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc4=accuracy_score(y_test, y_pred)
```

```python
print(f"Accuracy score:{acc4}")
```

# Using Decision Tree Classification Algo

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc5=accuracy_score(y_test, y_pred)
print(f"Accuracy score:{acc5}")
```

# Create visualization for model Accuracy

```python
mylist=[]
mylist2=[]
mylist.append(acc1)
mylist2.append("Logistic Regression")
mylist.append(acc2)
mylist2.append("SVM")
mylist.append(acc3)
mylist2.append("KNN")
mylist.append(acc4)
mylist2.append("Naive Bayes")
mylist.append(acc5)
mylist2.append("DTR")
```

```python
plt.rcParams['figure.figsize']=8,6
sns.set_style("darkgrid")
ax=sns.barplot(x=mylist2,y=mylist, palette="rocket", saturation=1.5)
plt.xlabel("Classifiction Models",fontsize=15)
plt.ylabel("Accuracy",fontsize=15)
plt.title("Accuracy of different Classifiction Models",fontsize=15)
plt.xticks(fontsize=11, horizontalalignment='center', rotation=6)
plt.yticks(fontsize=13)
for p in ax.patches:
    width, height= p.get_width(), p.get_height()
    x, y= p.get_xy()
    ax.annotate(f'{height:.2%}',(x + width/2, y + height*1.02), ha= 'center',fontsize='x-large' )
    plt.show()
```
